# isin method

```python
df= pd.read_csv('/Users/prashantsagar/Desktop/Data analysis/csvfile/employees.csv',parse_dates=["Start Date","Last Login Time"])
df.head()
```

```python
mask= df["First Name"].isin(["Jerry","Maria","Larry"])
df[mask]
# filter the given name from list
```

# isnull or notnull method

```python
mask2=df['Team'].isnull()
df[mask2]
# it shows all the null values in particular column
```

```python
df[df["Salary"].notnull()]
# df[mask3]
# it shows all the notnull values in particular column
```

# between method

```python
df[df['Salary'].between(90000,1000000)]
# it return the values botween given condition
```

```python
df[df["Start Date"].between("1998-02-11","2010-07-1-23")]
```

# duplicated method

```python
df[df['First Name'].duplicated()]
```

```python
df[df['First Name'].duplicated(keep='first')]
# in this only first duplicate name will be kept 
```

```python
df[df['First Name'].duplicated(keep='last')]
# in his last duplicated will be kept
```

```python
 df[df['First Name'].duplicated(keep= False)]
# it remove all duplicated values
# df[-df['First Name'].duplicated(keep= False)]  --viceversa

```
