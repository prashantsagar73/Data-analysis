# .apply method

```python
google.head(10)
```

```python
def performanc(number):
    if number<100:
        return "ok"
    elif number>100 and number<500:
        return "alright"
    elif number>700:
        return "supb"
    else:
        return "more data"
```

```python
google.apply(performanc).tail(3)
# apply methos use in selected data to filter it
```

```python
google.apply(lambda stock_price: stock_price+1).head(12)
# incres the value by give like 1
```
